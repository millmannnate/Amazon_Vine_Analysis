```python
import pandas as pd
```

```python
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
high_votes_df = vine_df.loc[vine_df['total_votes'] >= 20]
high_votes_df.head()
```

```python
helpful_votes_df = high_votes_df.loc[high_votes_df['helpful_votes']/high_votes_df['total_votes'] >= 0.5]
helpful_votes_df.head()
```

```python
yes_vine_df = helpful_votes_df.loc[helpful_votes_df['vine'] == 'Y']
yes_vine_df.head()
```

```python
no_vine_df = helpful_votes_df.loc[helpful_votes_df['vine'] == 'N']
no_vine_df.head()
```

```python
total_yes_reviews = yes_vine_df['review_id'].count()
print(total_yes_reviews)
yes_five_stars_df = yes_vine_df.loc[yes_vine_df['star_rating'] == 5]
yes_five_stars = yes_five_stars_df['review_id'].count()
print(yes_five_stars)
percentage_yes_five = (yes_five_stars/total_yes_reviews)*100
print(percentage_yes_five)
```

```python
total_no_reviews = no_vine_df['review_id'].count()
print(total_no_reviews)
no_five_stars_df = no_vine_df.loc[no_vine_df['star_rating'] == 5]
no_five_stars = no_five_stars_df['review_id'].count()
print(no_five_stars)
percentage_no_five = (no_five_stars/total_no_reviews)*100
print(percentage_no_five)
```

```python
summary_df = pd.DataFrame({
    "Vine":["Yes","No"],
    "Total reviews":[total_yes_reviews,total_no_reviews],
    "5 Star reviews":[yes_five_stars,no_five_stars],
    "Percentage 5 Stars":[percentage_yes_five,percentage_no_five]
})
summary_df
```
